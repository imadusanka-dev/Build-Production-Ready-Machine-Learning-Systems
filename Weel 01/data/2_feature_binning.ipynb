## 2. Basic Processing

```python
df = pd.read_csv('processed/ChurnModelling_Outliers_Handled.csv')
df.head()
```

350 - 500 -> 'Poor'
500 - 670 -> 'Fair'
670 - 740 -> 'Good'
740 - 800 -> 'Very Good'
800 - 850 -> 'Excellent'

```python
def custom_binning_credit_score(score):
    if score < 500:
        return 'Poor'
    if score < 670:
        return 'Fair'
    if score < 740:
        return 'Good'
    if score < 800:
        return 'Very Good'
    if score <= 850:
        return 'Excellent'
    else:
        assert True, "Credit Score cant't go beyon 850"

df['CreditScoreBins'] = df['CreditScore'].apply(custom_binning_credit_score)
del df['CreditScore']

df.head(10)
```

```python
df.to_csv('processed/ChurnModelling_Binning_Applied.csv', index=False)
```
